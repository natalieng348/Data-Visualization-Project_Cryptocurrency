## Import & Read data

```python
# import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read crypto prices data
crypto_daily_prices_df = pd.read_csv('crypto_daily_prices.csv')
crypto_daily_prices_df
```

```python
# read crypto returns data
crypto_daily_returns_df = pd.read_csv('crypto_daily_returns.csv')
crypto_daily_returns_df
```

## 1. Lineplot

Comparing the changes in BTC, ETH, and LTC prices during the period of 2014 to 2021 can give an idea of how these cryptocurrencies have performed relative to each other over time. 

One way to do this is by creating a line plot of the daily prices for each cryptocurrency, with all the plots on the same graph.

```python
# rename columns in crypto_daily_prices_df
new_col_names = {'BTC-USD Price': 'btc_usd_price', 
                 'ETH-USD Price': 'eth_usd_price', 
                 'LTC-USD Price': 'ltc_usd_price'}

crypto_daily_prices_df = crypto_daily_prices_df.rename(columns=new_col_names)
crypto_daily_prices_df
```

```python
# create the lineplot of all crypto prices
crypto_daily_prices_df.plot(x='Date', y=['btc_usd_price', 'eth_usd_price', 'ltc_usd_price'], figsize=(15,8), linewidth=2)

plt.title('Crypto Daily Prices 2014-2021')
plt.legend(loc='best')

plt.ylabel('Price ($)')
plt.xlabel('Date')
plt.grid(visible=True,linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

The line plot illustrates the volatility of daily BTC prices, which underwent a substantial increase during 2020 and 2021, possibly due to market sentiment and investor demand for the cryptocurrency.

## 2. Multiple Subplots

Creating multiple subplots of each cryptocurrency:

```python
# create side-by-side subplots
# set figure size
plt.figure(figsize = (15,15))
plt.suptitle('Cryptocurrency Daily Prices 2014-2021')

# btc subplot
plt.subplot(3, 1, 1)
crypto_daily_prices_df.plot(x='Date', y='btc_usd_price', kind='line', color='green', ax=plt.gca())
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.grid(visible=True, linestyle='--', linewidth=0.5)

# eth subplot
plt.subplot(3, 1, 2)
crypto_daily_prices_df.plot(x='Date', y='eth_usd_price', kind='line', color='orange', ax=plt.gca())
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.grid(visible=True, linestyle='--', linewidth=0.5)

# ltc subplot
plt.subplot(3, 1, 3)
crypto_daily_prices_df.plot(x='Date', y='ltc_usd_price', kind='line', color='blue', ax=plt.gca())
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.grid(visible=True, linestyle='--', linewidth=0.5)
```

The graphs reveal that until around June 2017, there were limited variations in the values of all observed cryptocurrencies. However, since then, there has been a marked increase in volatility in their prices. It can also be observed that while BTC and ETH have had an overall upward trend, LTC prices reached a peak in late 2017 and have remained relatively stable at lower levels for the rest of the period under consideration.

## 3. Scatterplots  

#### a. Scatterplot displaying BTC and ETH daily returns using Matplotlib:

```python
crypto_daily_returns_df
```

```python
# identify the x-axis and y-axis
X = crypto_daily_returns_df['BTC']
Y = crypto_daily_returns_df['ETH']

# create the scatterplot
from scipy.stats import linregress

plt.figure(figsize = (12,10))

# Add a masking feature to exclude 'NaN' values
mask = ~np.isnan(X) & ~np.isnan(Y)
plt.scatter(X[mask], Y[mask])

# Add x & y labels, title and grid
plt.xlabel('BTC daily returns')
plt.ylabel('ETH daily returns')
plt.title('BTC and ETH Daily Returns')
plt.grid(visible=True, linestyle='--', linewidth=0.5)

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(X[mask], Y[mask])

# Use the linear regression results to create the line
line_values = [slope*xi + intercept for xi in X[mask]]

# Add the line
plt.plot(X[mask], line_values, 'r', label='y={:.2f}x+{:.2f}, r={:.2f}'.format(slope, intercept, r_value))
plt.legend(loc='best')
plt.show()
```

The above graph suggests that there is a moderate positive correlation between BTC and ETH daily returns, with the correlation coefficient of 0.49. This indicates that as the daily return of BTC increases, the daily return of ETH also tends to increase. Similarly, as the daily return of BTC decreases, the daily return of ETH also tends to decrease.
 
It can also be observed that the two assets move in the same direction but the movements are not perfectly correlated. It's important to note that correlation does not imply causation, and other factors such as market sentiment, global economic conditions, and regulatory changes may be influencing the correlation.

#### b. Scatterplot displaying relationship between the daily returns of Bitcoin (BTC) and Litecoin (LTC) using Seaborn

```python
# create a scatter plot using seaborn
plt.figure(figsize = (12,10))
sns.scatterplot(x = crypto_daily_returns_df['BTC'], y = crypto_daily_returns_df['LTC'])
```

# 4. Piechart

Pie chart showing the portfolio allocation of XRP, BTC, LTC, ADA and ETH using Matplotlib.

Assumptions:

* We became bullish on XRP and decided to allocate 60% of our assets in it. 
* We also decided to equally divide the rest of our assets in other coins (BTC, LTC, ADA and ETH).
* Explode attribute to increase separation between BTC and the rest of the portfolio.

```python
# define a dictionary with all crypto allocation in a portfolio
allocation = {'Allocation %': [10, 10, 10, 60, 10]}
explode = (0, 0, 0, 0.2, 0)

portfolio_df = pd.DataFrame(data = allocation, index = ['BTC', 'ETH', 'LTC', 'XRP', 'ADA'])
portfolio_df.plot.pie(y = 'Allocation %', explode = explode, figsize = (10, 10), autopct='%1.1f%%')

plt.title('Crypto Portfolio Allocation', fontsize = 20)
plt.legend()
plt.show()
```

# 5. Histogram

Histogram of ETH daily returns using 30 bins with red color:

```python
# mean & stdev of distribution
mean = crypto_daily_returns_df['ETH'].mean()
stdev = crypto_daily_returns_df['ETH'].std()

# histogram
plt.figure(figsize = (20,15))
plt.hist(crypto_daily_returns_df['ETH'], bins = 30, alpha = 0.5, color = 'r')

plt.grid(visible=True, linestyle='--', linewidth=0.5)

plt.xlabel('Returns', fontsize = 25)
plt.ylabel('Frequency', fontsize = 25)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)

plt.title(f'ETH Returns Distribution: mean = {mean:.2f}, stdev = {stdev:.2f}', fontsize = 30)
plt.show()
```

# 6. Heatmap

Heatmap showing the correlation between all crypto daily returns using Seaborn:

```python
# create a heatmap
plt.figure(figsize = (12,10))
cm = crypto_daily_returns_df.corr()
sns.heatmap(cm, annot=True, cmap="Reds", annot_kws={"size": 12})

plt.title('Crypto Correlation Heatmap', fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize = 14)
plt.yticks(fontsize = 14)

plt.show()
```
